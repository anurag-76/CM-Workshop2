#Difference between np.dot() and np.matmul() is :
#np.dot()
1. np.dot(scalar, array) → works as scalar multiplication.
2. np.dot() → computes the inner product (sum of element‑wise products).

#np.matmul()
1. np.matmul(scalar, array) → raises an error (since scalars aren’t valid for matrix multiplication).
2. np.matmul(scalar, array) → raises an error (since scalars aren’t valid for matrix multiplication).

# USING quiver() function

```python
import numpy as np
import matplotlib.pyplot as plt
""" Problem 2: Create 2-dimensional position vectors and plot them. """
origin = np.array([0,0])
V = np.array([1,1])
W = np.array([-2.0,-1])
fig, ax = plt.subplots()
# Add vectors V and W to the plot
ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.text(origin[0]-0.1, origin[1]+0.1, "O")
ax.text(V[0], V[1], "v")
ax.text(W[0]-0.125, W[1]-0.1, "w")
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
U = np.array([1, 0.25 ,0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0,0,0, U[0],U[1],U[2], color='r', label='U')
ax.quiver(0,0,0, V[0],V[1],V[2], color='g', label='V')
ax.quiver(0,0,0, W[0],W[1],W[2], color='b', label='W')

ax.set_xlim([0,1]); ax.set_ylim([0,1]); ax.set_zlim([0,1])
ax.legend()
plt.show()
```

```python
U = np.array([-0.5, 0.5])
V = np.array([1, 1])
W = U + V   # vector addition

plt.quiver(0,0, U[0],U[1], angles='xy', scale_units='xy', scale=1, color='r', label='U')
plt.quiver(0,0, V[0],V[1], angles='xy', scale_units='xy', scale=1, color='g', label='V')
plt.quiver(0,0, W[0],W[1], angles='xy', scale_units='xy', scale=1, color='b', label='W')

plt.xlim(-1,2); plt.ylim(-1,2)
plt.grid(True); plt.legend()
plt.show()
```

```python
import numpy as np

def magnitude(vec):
    return np.sqrt(np.sum(vec**2))

U = np.array([1.5, 1.5, 2.0])
print("Magnitude of U:", magnitude(U))

```

```python
def direction_cosines(vec):
    mag = np.linalg.norm(vec)
    return vec / mag

O = np.array([1.5, 1.5, 2.0])
print("Direction cosines of O:", direction_cosines(O))
```

```python
def dot_product(a, b):
    return np.dot(a, b)

a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
print("Dot product:", dot_product(a, b))
```

```python
def cross_product(a, b):
    return np.cross(a, b)

a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
print("Cross product:", cross_product(a, b))
```
